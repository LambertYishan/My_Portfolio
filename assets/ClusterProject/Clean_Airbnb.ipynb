```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer

path = 'airbnb_listings_austin-1.csv'
df = pd.read_csv(path)

df.head()
```

```python
# total count of NaN values
df.info()
print(df.isnull().sum())
```

Processing data

```python
df.columns
```

```python
df.drop(['name', 'summary', 'space', 'description', 'experiences_offered','neighborhood_overview', 'notes', 'transit',
         'listing_url','host_id', 'host_name','host_location', 'host_about','neighbourhood', 'city','square_feet'], axis=1, inplace=True)

df.columns
```

```python
print(df.isnull().sum())
print(df['id'].count())
```

```python
df['security_deposit'] = df['security_deposit'].fillna(0)
nan_count = df['security_deposit'].isna().sum()
print(nan_count)

df['cleaning_fee'] = df['cleaning_fee'].fillna(0)
nan_count = df['security_deposit'].isna().sum()
print(nan_count)

df["weekly_price"] = df["weekly_price"].apply(lambda x: 1 if not pd.isna(x) else 0)
print(df.isnull().sum())
```

```python
df[['review_scores_rating',
       'review_scores_accuracy', 'review_scores_cleanliness',
       'review_scores_checkin', 'review_scores_communication',
       'review_scores_location', 'review_scores_value','host_response_rate',
       'bathrooms', 'beds','host_listings_count']] = df[['review_scores_rating',
       'review_scores_accuracy', 'review_scores_cleanliness',
       'review_scores_checkin', 'review_scores_communication',
       'review_scores_location', 'review_scores_value','host_response_rate',
       'bathrooms','beds','host_listings_count']].fillna(0)

df[['host_has_profile_pic','host_identity_verified',
    'host_is_superhost']] = df[['host_has_profile_pic',
                                'host_identity_verified','host_is_superhost']].fillna('f')

print(df.isnull().sum())
```

```python
print(df['host_response_time'].unique())
# Convert the 'response_time' column to dummy variables
dummies = pd.get_dummies(df["host_response_time"], prefix="response_time")

# Add the dummy variables back to the original dataframe
df = pd.concat([df, dummies], axis=1)
df.drop(['host_response_time'], axis=1, inplace=True)

print(df.isnull().sum())
```

```python
# Convert the 'response_time' column to dummy variables
dummies = pd.get_dummies(df["property_type"], prefix="property")

# Add the dummy variables back to the original dataframe
df = pd.concat([df, dummies], axis=1)
df.drop(['property_type'], axis=1, inplace=True)

df.columns
```

```python
# Convert the 'response_time' column to dummy variables
dummies = pd.get_dummies(df["room_type"], prefix="room")

# Add the dummy variables back to the original dataframe
df = pd.concat([df, dummies], axis=1)
df.drop(['room_type'], axis=1, inplace=True)

df.columns
```

```python
# Convert the 'response_time' column to dummy variables
dummies = pd.get_dummies(df["bed_type"], prefix="bed")

# Add the dummy variables back to the original dataframe
df = pd.concat([df, dummies], axis=1)
df.drop(['bed_type'], axis=1, inplace=True)

df.columns
```

```python
# Convert the 'response_time' column to dummy variables
dummies = pd.get_dummies(df['cancellation_policy'], prefix="cancellation")

# Add the dummy variables back to the original dataframe
df = pd.concat([df, dummies], axis=1)
df.drop(['cancellation_policy'], axis=1, inplace=True)

df.columns
```

```python
# Step 1: Clean and preprocess the 'amenities' column
df["amenities"] = df["amenities"].str.strip("{}""").str.split(",")

# Step 2: Use MultiLabelBinarizer to create dummy variables
mlb = MultiLabelBinarizer()
amenities_dummies = pd.DataFrame(mlb.fit_transform(df["amenities"]), columns=mlb.classes_)

# Step 3: Combine the original DataFrame with the dummy variables
df = pd.concat([df, amenities_dummies], axis=1)

# Optional: Drop the original 'amenities' column
df = df.drop(columns=["amenities"])

df.columns
```

```python
df = df.dropna(subset=["host_since"])
print(df.isnull().sum())
df.info()
```

```python
df['host_is_superhost'] = df['host_is_superhost'].apply(lambda x:1 if x == 't' else 0)
df['host_has_profile_pic'] = df['host_has_profile_pic'].apply(lambda x:1 if x == 't' else 0)
df['host_identity_verified'] = df['host_identity_verified'].apply(lambda x:1 if x == 't' else 0)
df['has_availability'] = df['has_availability'].apply(lambda x:1 if x == 't' else 0)
df['instant_bookable'] = df['instant_bookable'].apply(lambda x:1 if x == 't' else 0)
```

```python
df.to_csv('Cleaned_Airbnb.csv')
```
